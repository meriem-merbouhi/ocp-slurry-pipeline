```python
# Cellule 1 — Chemins
PATH_GOLD = "/Volumes/workspace/default/gold/"
```

```python
# Cellule 2 — Enregistrement des 5 tables
spark.sql("USE CATALOG workspace")
spark.sql("USE DATABASE default")

tables = ["fact_sensor_events", "dim_date", "dim_station", "dim_equipment", "dim_kpi"]

for table in tables:
    df = spark.read.parquet(f"{PATH_GOLD}{table}/")
    df.write.format("delta") \
      .mode("overwrite") \
      .saveAsTable(f"workspace.default.{table}")
    print(f"✅ {table} — {df.count()} lignes enregistrées")
```

```python
# Cellule 3 — Vérification
spark.sql("SHOW TABLES IN workspace.default").show()
```
